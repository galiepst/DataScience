# Feature Selection Strategy

```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
df = pd.read_csv("C:/Users/galie/DataScience/project/final_data_inc_newstuff.csv")
df.head()
```

```python
df['alot_of_patients_yn'] = df['alot_of_patients_yn'].astype('category')

df.dtypes
```

```python
# the first 3 ones are just indexes
varSel = pd.DataFrame({'Variable': df.columns[3:46]})
varSel.head()
```

### create a categorical variable for the checks below (as explained in class)

```python
nm = df.columns[3:46]
nm = nm.append(pd.Index(['alot_of_patients_yn']))
nm
```

```python
df.head()
```

```python
df2 = df[nm].copy()
df2.head()
```

## Multivariable Analysis

```python
### drop na
df2 = df2.dropna()
```

```python
## remove unnecessary vars
X = df2.iloc[:,:-1]
X.head()
```

```python
y = df2.iloc[:,-1:]
print([X.shape,y.shape])
```

### Variable Selection using LASSO (L1 penalization)

```python
from sklearn.linear_model import Lasso
from sklearn.feature_selection import SelectFromModel

lassomod = Lasso(alpha=0.01).fit(X, y)
```

```python
model = SelectFromModel(lassomod, prefit=True)
model.get_support()
```

```python
varSel['Lasso'] = model.get_support().astype('int64')
varSel
```

### Variable Selection using Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.feature_selection import SelectFromModel

rfmod = RandomForestClassifier().fit(X, y)
#rfmod.feature_importances_ 
```

```python
model = SelectFromModel(rfmod, prefit=True)
model.get_support()
```

```python
varSel['RandomForest'] = model.get_support().astype('int64')
varSel
```

### Variable Selection using Gradient Boosting classification

```python
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.feature_selection import SelectFromModel

gbmod = GradientBoostingClassifier().fit(X, y)
```

```python
model = SelectFromModel(gbmod, prefit=True)
model.get_support()
```

```python
varSel['GradientBoost'] = model.get_support().astype('int64')
varSel
```

### Variable Selection using SVM classification

```python
from sklearn.svm import LinearSVC
from sklearn.feature_selection import SelectFromModel

svmmod = LinearSVC(C=0.01, penalty="l1",dual=False).fit(X, y)
```

```python
model = SelectFromModel(svmmod, prefit=True)
model.get_support()
```

```python
varSel['SVM'] = model.get_support().astype('int64')
varSel
```

### Summarization and Selection of Variables 

```python
varSel['Sum'] =  np.sum(varSel,axis=1)
varSel
```

```python
varSel.groupby('Sum')['Variable'].count()
```
